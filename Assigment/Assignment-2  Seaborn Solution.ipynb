## 2: Plot 12 most dangerous cities

```python
kill.dropna(inplace=True)
```

```python
kill["city"].value_counts().head(12)
```

```python
# sns.set_theme()
sns.set_style('white')
a = kill["city"].value_counts().head(12)
plt.figure(figsize=(10,6))
sns.barplot(a.index ,  a.values,);
plt.xticks(rotation=45)
plt.title("Most dangerous cities", color="b");
```

```python
plt.figure(figsize=(10,6))
sns.countplot(x = "city"  ,data =kill, order =a.index);
plt.xticks(rotation=45);
plt.title("Most Dangereous City", color = "b", fontsize=14);
```
