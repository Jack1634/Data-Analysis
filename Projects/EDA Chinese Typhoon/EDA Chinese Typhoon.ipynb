```python
df["Time"] = pd.to_datetime(df["code"])
df["No"] = df["code"]//int(1e6)
```

```python
def concat_func(x):
    return pd.Series({
        "Start": x["Time"].min(),
        "End": x["Time"].max(),
        "StLat":x.loc[x["Time"].idxmin(),"LAT"],
        "EdLat":x.loc[x["Time"].idxmax(),"LAT"],
        "StLONG":x.loc[x["Time"].idxmin(),"LONG"],
        "EdLONG":x.loc[x["Time"].idxmax(),"LONG"],
        "WND":x["WND"].max(),
        "SiR34":x["SiR34"].max(),
        "SatSeR" :",".join(x["SATSer"].unique())

    })
```

```python
weather_data = df.groupby(df["No"]).apply(concat_func).reset_index()
```

```python
plt.figure(figsize=(15,5))
sns.kdeplot(weather_data["SiR34"],shade=True,color="r")
plt.title("Typhoon Size")
plt.show()
```

```python
plt.figure(figsize=(15,5))
sns.kdeplot(weather_data["WND"],shade=True,color="b")
plt.title("Typhoon Wind Strength")
plt.show()
```
