```python
from calendar import week
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

```

```python

# Plotly Express

import plotly.express as px

# create a dataframe
df = pd.DataFrame({"week" : np.random.randint(1,21,size=200),
"P1" : np.random.randint(10,220,size=200),
"P2" : np.random.randint(15,220,size=200),
"P3" : np.random.randint(10,490,size=200),
"P4" : np.random.randint(10,980,size=200)})

df = df.melt(id_vars=["week"], var_name="product", value_vars=df.columns[1:], value_name="sales")
df = df.groupby(["week", "product"]).sales.mean().reset_index()
df = df.sort_values(by=["week","product", "sales"])
```

```python
# Animated graphic

fig = px.bar(df, x="product", y="sales", color="product",
animation_frame="week", animation_group="product", range_y=[0,750])
fig.update_layout(height=600, width=800)
fig.show()
```

```python
fig =px.scatter(df, x ="week", y="sales", animation_frame="week",animation_group="product", color ="product",
hover_name="product",range_x = [0,20], range_y=[0,800])
fig.update_layout(height=600, width=1000)
```
