```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
 
```

```python
df = pd.read_csv("playoffStats.csv")
```

```python
df.head()
```

```python
df2 =df[df['season']==2022]
```

```python
df2.head()
```

```python
df2.groupby("team_id").pts_per_g.agg(["max","min","median","mean"])
```

```python
df2.groupby("team_id").pts_per_g.agg(["max","min","median","mean"]).style.format({"max": "{:.1f}", "min": "{:.1f}", "median": "{:.1f}", "mean": "{:.1f}"})
```

```python
df2.groupby("team_id").pts_per_g.agg(["max","min","median","mean"]).style.format( {**{col :"{:.1f}" for col in ["max","min","median","mean"]}}).set_caption("NBA Playoff  Points Per Game by Team")
```

```python
df2.groupby("team_id").pts_per_g.agg(["max","min","median","mean"]).style.highlight_max(axis=0, color="lightgreen").highlight_min(axis=0, color="lightcoral")
```

```python
# Please note that if you want to highlight by row, you need to specify axis=1

df2.groupby("team_id").pts_per_g.agg(["max","min","median","mean"]).style.highlight_max(axis=1, color="lightgreen").highlight_min(axis=1, color="lightcoral")
```

```python
# Let’s say we apply a gradient background color only to the ‘mean’ column:

df2.groupby("team_id").pts_per_g.agg(["max","min","median","mean"]).style.format( {**{col :"{:.1f}" for col in ["max","min","median","mean"]}}).set_caption("NBA Playoff  Points Per Game by Team").background_gradient(axis = 0, cmap='Blues', subset="mean",vmin=7.5, vmax=10)
```

```python
# We can also apply the gradients to the text with the text_gradient method:

df2.groupby("team_id").pts_per_g.agg(["max","min","median","mean"]).style.format( {**{col :"{:.1f}" for col in ["max","min","median","mean"]}}).set_caption("NBA Playoff  Points Per Game by Team").text_gradient(subset="mean", cmap='Blues', vmin=6.5, vmax=10)
```

```python
# Embedded bars

(df2.groupby('team_id')
 .obpm
 .agg([ 'mean'])
 .style
 .format({**{col: '{:.1f}' for col in ['max', 'min', 'median', 'mean']}})
 .bar(color=['lightcoral','lightgreen'],subset='mean',align='mid')
)
```
