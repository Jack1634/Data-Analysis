```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
labels = ['UK', 'USA', 'GER', 'JAP', 'FRA']
age_men = [40, 35, 45, 53, 28]
age_women = [35, 32, 48, 48, 31]

plt.bar(labels,age_men, width=0.35, label="Men"  )
plt.title("Age means by Gender")
plt.ylabel("Ages")
plt.legend(loc=1)
p = range(len(labels))

for i in range(len(labels)):
    
    plt.text(x = i, y = age_men[i], s = age_men[i], ha="center", size = 10,va="bottom")
plt.show()
```

```python
labels = ['UK', 'USA', 'GER', 'JAP', 'FRA']
age_men = [40, 35, 45, 53, 28]
age_women = [35, 32, 48, 48, 31]

plt.bar(labels,age_women, width=0.35, label="Women"  )
plt.title("Age means by Gender")
plt.ylabel("Ages")
plt.legend(loc=1)
p = range(len(labels))

for i in range(len(labels)):
    
    plt.text(x = p[i], y = age_women[i], s = age_women[i], ha="center", size = 10, va="bottom")
plt.show()
```

```python
fig, ax = plt.subplots()
ax.bar(labels,age_men, label="Men")
ax.set_title("Age means by Gender")
ax.set_ylabel("Ages")
plt.legend()
for p in ax.patches:
    ax.annotate((p.get_height()),(p.get_x()+0.35,p.get_height()+0.15));
```

```python
xpos=np.arange(len(labels))
plt.bar(xpos-0.2,age_men, width=0.35, label="Men"  )

plt.bar(xpos+0.2,age_women, width=0.35, label="Women")

for i in range(len(labels)):
    
    plt.text(x = i-0.2, y = age_men[i], s = age_men[i], ha="center", size = 10, va="bottom")
    
for i in range(len(labels)):
    
    plt.text(x = i+0.2, y = age_women[i], s = age_women[i], ha="center", size = 10, va="bottom")
    
plt.xticks(xpos,labels)

plt.title("Age means by Gender")
plt.ylabel("Ages")
plt.legend()
plt.show()
```
