```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import pandas as pd
import seaborn as sns

import warnings;
warnings.filterwarnings('ignore')
```

```python
age = [25, 27, 29, 31, 33, 35, 37, 39, 41, 43, 45]

salary = [38496, 42000, 46752, 49320, 53200,
         56000, 62316, 64928, 67317, 68748, 73752]

salary_2 = [45372, 48876, 53850, 57287, 63016,
            65998, 70003, 70000, 71496, 75370, 83640]
```

# Review

```python
# Functional 
plt.plot(age, salary)
plt.xlabel("age")
plt.ylabel("salary")
plt.title("Salary by Age");
plt.show()
```

```python
# Object Oriented
fig, ax = plt.subplots(figsize=(6,3))
ax.plot(age, salary, 'r')
ax.set_xlabel('age')
ax.set_ylabel('salary')
ax.set_title('title')

plt.show()
```

```python
plt.plot(age,salary,age,salary_2)

plt.show()
```

Subplot

```python
plt.figure(figsize=(15,3))

plt.subplot(1,2,1)
plt.plot(age, salary, "r+", ls="--")


plt.subplot(1,2,2)
plt.plot(age, salary_2, "bo", ls=":")

plt.show()
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(15,3))

ax[0].plot(age, salary)
ax[0].set_xlabel('age')
ax[0].set_title('First')

ax[1].plot(age, salary_2)
ax[1].set_title('Second')
ax[1].set_xlabel('age')
plt.tight_layout()

plt.show()
```

```python
tips = sns.load_dataset("tips")
tips.head()
```

```python
tips.groupby("day").sum()
```

```python
day = tips.groupby("day").sum().index   # tips.day.unique()
day
```

```python
tips.day.unique()
```

```python
day_of_total_bill= list(tips.groupby("day")["total_bill"].sum())
day_of_total_bill
```

```python
fig, ax = plt.subplots(figsize=(15,5))

ax.barh(day,day_of_total_bill)
plt.show()
```

```python
day_tip= np.array(tips.groupby("day")["tip"].sum())
day_tip
```

```python
fig, ax = plt.subplots(figsize=(10,5))


ax.bar(day, day_of_total_bill,  label="total_bill")
ax.bar(day, day_tip,            label="tip")

plt.legend()
```

### Optional

```python
# Optional hard to coding

fig, ax = plt.subplots(figsize=(10,5))

p = np.arange(len(day))
width = 0.15

ax.bar(p - width/2, day_of_total_bill, width, label="total_bill")
ax.bar(p + width/2, day_tip,           width, label="tip")

ax.set_xticks(p)
ax.set_xticklabels(day)

plt.legend()
plt.show()
```

# Pie Charts

```python
plt.figure(figsize=(10,8))
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels, labeldistance=0.7, autopct="%.1f")
plt.show()
```

```python
plt.figure(figsize=(10,8))
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0.1, 0, 0, 0]

plt.pie(y, labels = mylabels, explode = myexplode,labeldistance=0.7, autopct="%.1f")
plt.legend()
plt.show()
```

# Three-Dimensional Plotting in Matplotlib

```python
from mpl_toolkits import mplot3d
fig = plt.figure()
ax = plt.axes(projection='3d')
```

```python
ax = plt.axes(projection='3d')

# Data for a three-dimensional line
zline = np.linspace(0, 15, 1000)
xline = np.sin(zline)
yline = np.cos(zline)
ax.plot3D(xline, yline, zline, 'gray')
```

```python
ax = plt.axes(projection='3d')
# Data for three-dimensional scattered points
zdata = 15 * np.random.random(100)
xdata = np.sin(zdata) + 0.1 * np.random.randn(100)
ydata = np.cos(zdata) + 0.1 * np.random.randn(100)
ax.scatter3D(xdata, ydata, zdata, c=zdata, cmap='Reds');
```
